```python
dataset = pd.read_csv('skill2vec_50K.csv', header = 0, encoding='ISO-8859-1', low_memory=False)
dataset.head()
```

```python
data = []
for index, row in tqdm(dataset.iterrows()):
    job_brief_id = row[0]
    keyword_name = row[1]
    job_titles = []
#     print(job_brief_id)
#     print(keyword_name)
    
    row = row.drop(row.index[0]) # remove job_brief_id
    row = row.drop(row.index[0]) # remove keyword_name
    for item in row:
        if str(item) != "nan":
            curr_row = []
            curr_row.append(job_brief_id)
            curr_row.append(keyword_name)
            curr_row.append(item)
            data.append(curr_row)
```

```python
formatted_data = pd.DataFrame(data, columns = ['job_brief_id', 'keyword_name', 'job_title'])
formatted_data.head()
```
